```python
import pandas as pd
import numpy as nm
import seaborn as sns
import matplotlib.pyplot as plt
import statsmodels.api as sn
```

```python
df = pd.read_csv("G:\Online Trainings\Letsupgrade\Day 21\general_data.csv")
df.head()
```

```python
df.dropna(inplace=True)
df.shape
```

```python
df = df.drop(columns = ['EmployeeID', 'EmployeeCount', 'Over18', 'StandardHours', ])
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
enc = LabelEncoder()
df.Attrition = enc.fit_transform(df.Attrition)
df['BusinessTravel'] = enc.fit_transform(df['BusinessTravel'])
df['Department'] = enc.fit_transform(df['Department'])
df['EducationField'] = enc.fit_transform(df['EducationField'])
df['Gender'] = enc.fit_transform(df['Gender'])
df['JobRole'] = enc.fit_transform(df['JobRole'])
df['MaritalStatus'] = enc.fit_transform(df['MaritalStatus'])
df
```

```python
df.info()
```

```python
y = df['Attrition']
x = df.drop(columns = ['Attrition'])
```

```python
X = sn.add_constant(x)
logistic_model1 = sn.Logit(y,X)
```

```python
result1 = logistic_model1.fit()
```
